# Doubly Linked Lists

- It is another **data structure** â€” so the aim is to be able to store items more efficiently (insertion and removal operations)
- The disadvantage that arrays have is the holes occuring when we remove an element.
- This problem can be eliminated through **doubly linked lists.**

- Every node stores the **data** itself and the **references** to the next node and to the previous node in the linked list.
- That's why **doubly linked lists** need more memory than singly linked lists.
- It has an advantage â€” there cannot be *holes* in the data structure so there is **no need for shifting items.**

### Advantages of Doubly Linked Lists

1. We store references to the **head node and the tail node**, so these nodes can be accessed in ***O(1)*** constant running time. 
2. It can be traversed in both directions.
3. Removing a given node is easier because there is a pointer because there is a pointer to the **previous node** as well.

### Disadvantages of Doubly Linked Lists

1. Need **more memory** because of the references (2 pointers)
2. A bit more **complicated** to implement because we have to handle both of the pointers.
3. Still have not solved the main issue â€” how to seearch for arbitrary elements faster than ***O(N)*** linear running time?
